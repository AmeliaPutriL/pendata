## Perhitungan Manual

### 2. Mengambil contoh threshold split :

* Threshold = 212.5

Bagi kategori:
* Group Left (≤ 212.5)
* Group Right (>212.5)

### 3. Menghitung Entropy Awal (Sebelum split)

$$
Entropy(D) = -p_1 \log_2(p_1) - p_2 \log_2(p_2) - p_3 \log_2(p_3)
$$

$$
p_1 = \frac{7}{15},\quad p_2 = \frac{5}{15}, \quad p_3 = \frac{3}{15}
$$

$$
Entropy(D) = -\frac{7}{15} \log_2 \frac{7}{15} + \frac{5}{15} \log_2 \frac{5}{15} + \frac{3}{15} \log_2 \frac{3}{15}\approx 1.503
$$

### 4. Entropy setelah split

Group Left (≤ 212.5)

* Data: No 1-7
* Kelas: 3 Panthera uncia, 4 Panthera pardus

$$
Entropy(left) = -\frac{3}{7}\log_2\frac{3}{7} - \frac{4}{7}\log_2\frac{4}{7} \approx 0.985
$$

Group Right (> 212.5):

* Data: No 8 - 15
* Kelas: 7 Panthera onca, 1 Panthera pardus

$$
Entropy(right) = -\frac{7}{8} \log_2 \frac{7}{8} -\frac{1}{8} \log_2 \frac{1}{8} ≈ 0.543
$$

Information Gain:

$$
IG = Entropy(D) - \left( \frac{7}{15} \cdot 0.985 + \frac{8}{15} \cdot 0.543 \right) = 1.503 - (0.459 + 0.289)\approx \boxed{0.755}
$$

### 5. Gini Index

$$
Gini(D) = 1 - \left(\left(\frac{7}{15}\right)^2 + \left(\frac{5}{15}\right)^2 + \left(\frac{3}{15}\right)^2\right) \approx 0.631
$$

Group Left (≤ 265):

* 3 Panthera uncia, 4 Panthera pardus

$$
Gini_{left} = 1 - \left(\frac{3}{7}\right)^2 - \left(\frac{4}{7}\right)^2 = 0.489
$$

Group Right (> 212.5):

* 7 Panthera onca, 1 Panthera pardus

$$
Gini_{right} = 1 - \left(\frac{7}{8}\right)^2 - \left(\frac{1}{8}\right)^2 = 0.218
$$

Gini Split:

$$
Gini_{split} = \frac{7}{15} \cdot 0.489 + \frac{8}{15} \cdot 0.218 \approx 0.345
$$

### Hasil Sementara untuk Fitur "Panjang Tubuh":

### 2. Ambil Threshold Split

Gunakan threshold = **75 kg**

- **Group Left (≤ 75)**: No 14–8 → 10 data
- **Group Right (> 75)**: No 3–4 → 5 data



### 3. Komposisi Kelas Sebelum Split

- 5 *Panthera onca*  
- 5 *Panthera pardus*  
- 5 *Panthera uncia*

$$
Entropy(D) = -\frac{5}{15} \log_2 \frac{5}{15} \times 3 = \log_2 3 \approx \boxed{1.585}
$$



### 4. Entropy Setelah Split

#### Group Left (≤ 75 kg)

- 5 *Panthera uncia*
- 5 *Panthera pardus*

$$
Entropy(left) = -\left( \frac{5}{10} \log_2 \frac{5}{10} + \frac{5}{10} \log_2 \frac{5}{10} \right) = \boxed{1}
$$

#### Group Right (> 75 kg)

- 5 *Panthera onca*

$$
Entropy(right) = -1 \cdot \log_2 1 = \boxed{0}
$$



### 5. Information Gain

$$
IG = Entropy(D) - \left( \frac{10}{15} \cdot 1 + \frac{5}{15} \cdot 0 \right) = 1.585 - 0.6667 = \boxed{0.918}
$$



### 6. Gini Index

#### Gini Awal:

$$
Gini(D) = 1 - \left(\frac{1}{3}\right)^2 \cdot 3 = 1 - \frac{1}{3} = \boxed{0.667}
$$

#### Gini Left (≤ 75 kg)

- 5 uncia, 5 pardus

$$
Gini(left) = 1 - \left( \frac{5}{10} \right)^2 - \left( \frac{5}{10} \right)^2 = 1 - 0.25 - 0.25 = \boxed{0.5}
$$

#### Gini Right (> 75 kg)

- Semua 5 *Panthera onca* → satu kelas

$$
Gini(right) = 0
$$

#### Gini Split:

$$
Gini_{split} = \frac{10}{15} \cdot 0.5 + \frac{5}{15} \cdot 0 = \boxed{0.333}
$$

## Visualisasi Decision Tree

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data ===
data = {
    "Panjang Tubuh (cm)": [240, 250, 230, 260, 255, 210, 200, 220, 215, 205, 180, 190, 195, 185, 200],
    "Berat Tubuh (kg)": [100, 110, 95, 120, 115, 65, 60, 70, 68, 62, 50, 55, 52, 48, 58],
    "Kelas": [
        "Panthera onca", "Panthera onca", "Panthera onca", "Panthera onca", "Panthera onca",
        "Panthera pardus", "Panthera pardus", "Panthera pardus", "Panthera pardus", "Panthera pardus",
        "Panthera uncia", "Panthera uncia", "Panthera uncia", "Panthera uncia", "Panthera uncia"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # encoding otomatis: uncia/pardus/onca → angka

# === Fitur dan Target ===
X = df[["Panjang Tubuh (cm)", "Berat Tubuh (kg)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)
```

## Visualisasi Splitting

```python
# === Uji 2 Data Baru ===
uji_baru = pd.DataFrame({
    'Panjang Tubuh (cm)': [235, 248],
    'Berat Tubuh (kg)': [105, 112]  # Sesuaikan dengan rentang berat pada data awal
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (p, b) in enumerate(zip(uji_baru['Panjang Tubuh (cm)'], uji_baru['Berat Tubuh (kg)'])):
    print(f"Data {i+1}: Panjang={p}, Berat={b} → Prediksi: {hasil[i]}")

# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
from matplotlib.colors import ListedColormap

# Warna dan label untuk 3 kelas
colors = ["#FFAAAA", "#AAFFAA", "#AAAAFF"]
cmap = ListedColormap(colors[:len(le.classes_)])

x_min, x_max = X['Panjang Tubuh (cm)'].min() - 10, X['Panjang Tubuh (cm)'].max() + 10
y_min, y_max = X['Berat Tubuh (kg)'].min() - 10, X['Berat Tubuh (kg)'].max() + 10

xx, yy = np.meshgrid(np.arange(x_min, x_max, 1.0),
                     np.arange(y_min, y_max, 1.0))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Panjang Tubuh (cm)', 'Berat Tubuh (kg)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.6)

# Titik data training
for label in np.unique(y):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Panjang Tubuh (cm)'], X.iloc[idx]['Berat Tubuh (kg)'],
                edgecolor='k', label=le.inverse_transform([label])[0])

# Titik uji baru
plt.scatter(uji_baru['Panjang Tubuh (cm)'], uji_baru['Berat Tubuh (kg)'],
            c='black', marker='x', s=100, label='Data Uji')

plt.xlabel("Panjang Tubuh (cm)")
plt.ylabel("Berat Tubuh (kg)")
plt.title("Decision Boundary (Panjang vs Berat)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualisasi Splitting Setelah ditambahkan 2 Data Baru

```python
# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")

plot_colors = ["r", "y", "b"]  # merah, kuning, biru
plot_step = 1.0

x_min, x_max = X['Panjang Tubuh (cm)'].min() - 10, X['Panjang Tubuh (cm)'].max() + 10
y_min, y_max = X['Berat Tubuh (kg)'].min() - 10, X['Berat Tubuh (kg)'].max() + 10

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Panjang Tubuh (cm)', 'Berat Tubuh (kg)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training untuk 3 label
for label, color in zip(np.unique(y), plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Panjang Tubuh (cm)'], X.iloc[idx]['Berat Tubuh (kg)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

# Plot data uji
plt.scatter(uji_baru['Panjang Tubuh (cm)'], uji_baru['Berat Tubuh (kg)'],
            c='black', marker='X', s=100, label='Data Uji')

plt.xlabel("Panjang Tubuh (cm)")
plt.ylabel("Berat Tubuh (kg)")
plt.title("Decision Boundary (Panjang vs Berat)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualisasi Decision Tree dalam Bentuk Gambar

```python
# === Visualisasi Struktur Pohon ===
plt.figure(figsize=(12, 3))
plot_tree(
    clf,
    feature_names=["Panjang Tubuh (cm)", "Berat Tubuh (kg)"],
    class_names=le.classes_,
    filled=True,
    rounded=True,
    impurity=True  # tampilkan nilai Gini
)
plt.title("Decision Tree untuk Panthera onca vs pardus vs uncia")
plt.tight_layout()
plt.show()
```
